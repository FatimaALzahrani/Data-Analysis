## Sales prediction


The project aims to forecast a perfume shop's sales in order to optimize inventory management. The shop conducts events both in-store and on social media. Sales are notably influenced by seasonal variations and weather conditions. The store can easily provide its sales data and the planning of its physical events. If additionnal features are needed, we will have to get them by ourselves.



### Questions


What are the features and what is the target?


> **The target is the store sales. The already identified features are social network events, weather conditions, seasonality (day, month, holidays etc.)**


Are we talking about supervised or unsupervised learning?



> **In this use case, we have an identified target (the sales) and various features (seasonality, weather conditions etc.) associated to the target. Both elements will be used during the training of the ML model, so we are dealing with supervised learning.**




Is the data structured or unstructured?


> **Data is structured with dates in rows and features in columns**



Is it a regression or classification problem? Neither?


> **It is a regression problem because the store sales are continuous real values**


## Detection of default risk


The project aims is to detect if a person is at risk of defaulting on a loan. This will save time to focus only on the most critical cases. Both loan and applicant data are already in the database.

### Questions


What are the features and what is the target?


> **The target is the loan repayment status, and the features are all the loan and applicant data**



Are we talking about supervised or unsupervised learning?



> **In this use case, we have an identified target (the repayment status) and various features (loan and applicant data) associated to the target. Both elements will be used during the training of the ML model, so we are dealing with supervised learning.**



Is the data structured or unstructured?


> **The data is structured with loan requests in rows and features in columns**



Is it a regression or classification problem? Neither?


> **It is a binary classification problem, since a loan can either be at repaid or not.**



## Customer segmentation


The company specializes in delivering sustainable baskets. The project aims to enhance understanding of their customers by segmenting them based on their characteristics. This segmentation will allow the company to redesign the customer experience specifically for each customer profile. The company has provided their basket sales data, and we already asked for any additional information they may have on their customers (in progress).

### Questions


What are the features and what is the target?


> **There is no target, we will simply use features to create customers groups.**


Are we talking about supervised or unsupervised learning?



> **Since we do not have a specific target, we will only use our features during the training of the ML model, which will be used to segment our databse. Therefore, we are dealing with unsupervised learning.**

Is the data structured or unstructured?


> **We have not yet received all the data, but it is more likely to be structured data. We have customer data displayed in rows, with customer features in columns.**


Is it a regression or classification problem? Neither?


> **Neither, it is a clustering problem.**


## Fracture diagnostics

The company is a radiology practice, and it wants to implement a fracture type detection algorithm. With seven distinct fracture types (transverse, longitudinal, displaced oblique, non-displaced oblique, greenwood, comminuted, spiral), accurate identification can be challenging, especially for rare cases. To address this, the goal is to develop an algorithm based on a comprehensive dataset of X-rays and their confirmed diagnoses.

### Questions


What are the features and what is the target?


> **The target is the type of fractures (labelled from 0 to 6). The features are the details presented in the X-rays.**


Are we talking about supervised or unsupervised learning?



> **In this use case, we have an identified target (the type of fracture) and various features (visual elements on the X-rays) associated to the target. Both elements will be used during the training of the ML model, so we are dealing with supervised learning.**


Is the data structured or unstructured?


> **Data is not structured, as we only have X-rays(which are images).**


Is it a regression or classification problem? Neither?


> **It is a multi-class classification problem, since we have 7 different types of fractures that we could detect.**
