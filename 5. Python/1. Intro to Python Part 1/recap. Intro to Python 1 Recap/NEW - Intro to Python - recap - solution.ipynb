# Functions

Return the sum of all the positive numbers using a *for* loop and an *if* condition.

```python
numbers = [3, -8, 46, 73, -53, 22, -90, 5, 78]
```

```python
# this is one way to do it, there are many other
def sum_of_positive(numbers):
  i=0
  for number in numbers:
    if number>0:
      i = i + number
  return i
```

```python
sum_of_positive(numbers)
```
